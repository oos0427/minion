# 합성곱 신경망의 구성 요소

합성곱 : 입력 데이터에 마법의 도장(가중치)를 찍어서 유용한 특성만 드러나게 하는 것으로 비유

특성맵 : 합성곱 계산을 통해 얻은 출력을 특별히 특성 맵이라고 부릅니다.
  * (4, 4) 크기의 입력을 (3, 3) 크기의 합성곱을 수행하면 (2, 2) 크기의 출력으로 압축된 느낌이 난다.

패딩 : 입력배열의 주위를 가상의 원소(0)로 채우는 것을 패딩이라고 합니다.

세임패딩 : 44크기의 입력을 44크기의 특성 맵으로 만들기 위해 44입력특성 주위에 0으로 채워 66크기의 입력값으로 만들어 특성 맵을 44크기로 만듬
-> 입력 값과 특성맵의 크기를 같게 하기위해 입력값 주위에 0을 채워 크기를 키움

밸리드 패딩 : 순수한 입력 배열에서만 합성곲을 하여 특성 맵을 만듬

풀링 : 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역활을 수행
       특성맵에서 최대값 또는 평균값을 뽑아 특성맵을 줄임
       최대 풀링, 평균 풀링
