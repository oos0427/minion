```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks(np.linspace(-np.pi, np.pi, 4))
#plt.xticks([-np.pi, -np.pi/2, 0,
#np.pi/2, np.pi])
plt.yticks([-1, 0, 1],
["Low", "Zero", "High"])
plt.grid(False)
plt.show()
```

```python
# 
# ax1 = plt.subplot(2, 1, 1)  2 1     
# subplots .
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)
ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)
plt.show()
```
