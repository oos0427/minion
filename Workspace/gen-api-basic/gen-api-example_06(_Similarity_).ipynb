```python
import pandas as pd
data = [
    "이제 집에 가고 싶다.",
    "배가 고파서 저녁을 빨리 먹고 싶다.",
    "이러다가 굶어서 죽을 것 같다.",
    "퇴근 시간이 가까워 지고 있다.",
    "트랜스포머는 시퀀스-시퀀스 구현 모델입니다.",
    "저녁으로는 쭈꾸미볶음이 먹고 싶다."
]

df = pd.DataFrame(data, columns=['text'])
df
```

```python
df['embedding'] = df.apply(lambda x: embeddings.embed_query(x["text"]), axis=1)
```

```python
df
```

```python
import numpy as np
def cos_sim(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))

query = "맛있는 음식을 먹는 것은 언재나 즐겁습니다."

embedded_query = embeddings.embed_query(query)
df['similarity'] = df['embedding'].map(lambda x: cos_sim(x, embedded_query))
df.sort_values(by='similarity', ascending=False)
```
