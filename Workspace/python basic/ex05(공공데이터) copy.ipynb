```python
#print(dict_response.keys())
#print(dict_response['response'].keys())
#print(dict_response['response']['body'].keys())
#print(dict_response['response']['body']['items'].keys())
print(dict_response['response']['body']['items']['item'])
```

```python
#   
minta = []
maxta = []
avgta = []
dates = []
for item in dict_response['response']['body']['items']['item']:
    #print(item['maxTa'], item['minTa'], item['avgTa'])
    minta.append(float(item['minTa']))
    maxta.append(float(item['maxTa']))
    avgta.append(float(item['avgTa']))
    dates.append(item['tm'])
```

```python
print(minta)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 3))
plt.plot(dates, avgta, label='avg')
plt.plot(dates, maxta, label='max')
plt.plot(dates, minta, label='min')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
