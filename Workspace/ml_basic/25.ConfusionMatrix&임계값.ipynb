```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
#데이터 준비
from sklearn.datasets import load_digits

data = load_digits()
np.unique(data['target'], return_counts= True)
```

불균형데이터세트 만들기

```python
x = data['data']
y = data['target']
#yis3 = (y==3), True, false 불린형태라 머신러닝에 사용 불가
yis3 = (y==3).astype(int)
```

```python
print(np.unique(yis3, return_counts=True))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, yis3, random_state=777)
```

```python
#모델훈련
from sklearn.dummy import DummyClassifier #알고리즘으로 학습하지 않는 모델
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(x_train, y_train)
logreg = LogisticRegression(max_iter=10000)
logreg.fit(x_train, y_train)
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
# 정확도 기반 모델 평가의 문제점
print(np.round(dummy.score(x_test, y_test), 2),
      np.round(logreg.score(x_test, y_test)), 
      np.round(dtc.score(x_test, y_test)))
```

```python
# cofusion matrix 계산 및 표시
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm1  = confusion_matrix(y_test, dummy.predict(x_test))
cm2  = confusion_matrix(y_test, logreg.predict(x_test))

print(cm1)
print(cm2)
#알고 싶은거에 1(True)를 지정해야한다.
ConfusionMatrixDisplay.from_estimator(logreg, x_test, y_test, display_labels=['3', 'not 3'])
```

조화평균 = 정밀도 * 재현율 / (정밀도 + 재현율)

```python
# 모델평가2
from sklearn.metrics import recall_score, precision_score, f1_score

print( recall_score(y_test, dummy.predict(x_test)),
       precision_score(y_test, dummy.predict(x_test)),
       f1_score(y_test, dummy.predict(x_test)))

print( recall_score(y_test, logreg.predict(x_test)),
       precision_score(y_test, logreg.predict(x_test)),
       f1_score(y_test, logreg.predict(x_test)))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, dummy.predict(x_test)))
print(classification_report(y_test, logreg.predict(x_test)))
```

```python
tdv = logreg.decision_function(x_train)
print(tdv[tdv > 10].shape)
print(tdv[tdv < -10].shape)

tpv = logreg.predict_proba(x_train)
tv = tpv[:][:, 1]
#np.round(tpv[ tv > 0.5][:, 1], 2)
print(tpv[tv > 0.7].shape)
print(tpv[tv < 0.3].shape)
```

```python
#데이터준비
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=(4000,500), cluster_std=[7.0, 2], random_state=42)
x_trian, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=42)
print(x.shape)

np.unique(y, return_counts=True)
```

```python
#모델훈련
from sklearn.svm import SVC
svc = SVC(gamma=0.5)

svc.fit(x_trian, y_train)
```

```python
from sklearn.metrics import precision_recall_curve

# 모델 예측에 따라 정밀도, 재현율, 임계값 계산; 임계치 변경에 따른 변화율
precisions, recalls, thresholds = precision_recall_curve(y_test, svc.decision_function(x_test))

# thresholds 확인
print("Thresholds:", thresholds)

idx = np.argmin(np.abs(thresholds)) # 0에 가장 가까운 threshold값

plt.figure(figsize=(10,4))
plt.plot(precisions, recalls, label = 'precisions_recall_curve')
plt.plot(precisions[idx], recalls[idx], 'o', markersize=10,label = 'threshold 0')
plt.xlabel('precision')
plt.ylabel('recall')
plt.legend()
plt.show()
idx
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, svc.decision_function(x_test))

idx = np.argmin(np.abs(thresholds)) # 0에 가장 가까운 threshold값

plt.figure(figsize=(10,4))
plt.plot(fpr, tpr, label = 'ROC-curve')
plt.plot(fpr[idx], tpr[idx], 'o', markersize=10,label = 'threshold 0', fillstyle= 'none')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.legend()
plt.show()
idx
```

```python
from sklearn.metrics import roc_auc_score

auc = roc_auc_score(y_test, svc.decision_function(x_test))

auc
```
