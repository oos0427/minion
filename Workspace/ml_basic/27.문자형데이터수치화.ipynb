```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#오늘 날씨는 맑고 기온은 3도 입니다.
#오늘 점심은 제육볶음을 먹을 생각입니다.
```

```python
from sklearn.datasets import load_files
```

```python
train_data = load_files('data/hide_data/train')
test_data = load_files('data/hide_data/test')
```

```python
#기본정보탐색
print(train_data.keys())
print(len(train_data['data']), len(test_data['data']))
print(np.unique(train_data['target'], return_counts=True))
print( train_data['target_names'])
print( train_data['data'][0])
```

```python
#효과적인 전처리를 위해 DataFrame으로 변경
train_df = pd.DataFrame({'label': train_data['target'], 'review' : train_data['data']})
test_df = pd.DataFrame({'label': test_data['target'], 'review' : test_data['data']})
```

```python
# binary data -> char text
train_df['review'] = train_df['review'].str.decode(encoding='utf-8')
test_df['review'] = test_df['review'].str.decode(encoding='utf-8')
```

```python
# decode  확인 및 저장
train_df.to_csv('data/imdb_reviews_train.csv', index = False)
test_df.to_csv('data/imdb_reviews_test.csv', index = False)
```

```python
train_df.info()
```

```python
train_df['review'] = train_df['review'].str.lower()
test_df['review'] = test_df['review'].str.lower()
```

```python
# review에 등록된 마크업(tag)가 포함된 것 확인 : <br />
#train_df['review'][0]
#train_df['review'] = train_df['review'].str.replace('<br / >', '')
from bs4 import BeautifulSoup

# BeautifulSoup("<div>test</div>").get_text() # test만 추출

train_df['review'] = train_df['review'].map(lambda v: BeautifulSoup(v, 'html.parser').get_text())
test_df['review'] = test_df['review'].map(lambda v: BeautifulSoup(v, 'html.parser').get_text())
```

```python
train_df['review'][0]
```

```python
#전처리 3 : 영문자만 남기기 (특수문자 제거)
import re
train_df['review'] = train_df['review'].map(lambda v : re.sub('[^A-za-z]', ' ', v))
test_df['review'] = test_df['review'].map(lambda v : re.sub('[^A-za-z]', ' ', v))
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#전처리 4 : 문장(string) -> 단어(토큰)의 리스트로 변경 (string list)
import nltk # 자연어 처리를 위한 패키지
train_df['review'] = train_df['review'].map(nltk.word_tokenize)
test_df['review'] = test_df['review'].map(nltk.word_tokenize)
```

```python
train_df['review'][0]
```

```python
#전처리 5 : 불용어(stopwords) 제거
#nltk.download('stopwords')
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')

print(en_stopwords[:10])

#en_stopwords += ['test', 'xyz'] 기존 stopwords 새단어 추가

train_df['review'] = train_df['review'].map(lambda v: [token for token in v if token not in en_stopwords])
test_df['review'] = test_df['review'].map(lambda v: [token for token in v if token not in en_stopwords])
```

```python
# 전처리 6 : 어근 추출(단어 활용에 의해 철자가 달라진 것을 표준화 또는 통합)
from nltk import PorterStemmer

stemmer = PorterStemmer()
# print(stemmer.stem('happy'), stemmer.stem('apples'), stemmer.stem('apple'))

train_df['review'] = train_df['review'].map(lambda v: [stemmer.stem(w) for w in v])
test_df['review'] = test_df['review'].map(lambda v: [stemmer.stem(w) for w in v])
```

```python
#전처리 7 : 토큰리스트를 단일 문장(single string)으로 변환
train_df['review'] = train_df['review'].map(lambda v : ' '.join(v))
test_df['review'] = test_df['review'].map(lambda v : ' '.join(v))
```

```python
# 전처리 8 : token -> numeric value 토큰의 숫자화
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
sparse_train_counts = cv.fit_transform(train_df['review'])
sparse_test_counts = cv.transform(test_df['review']) # 테스트는 훈련 없이 훈련한 모델을 적용
```

```python
# print(sparse_train_counts)
print( "단어사전의 크기 : {0}".format(len(cv.vocabulary_.keys())))
idx_to_word = { idx : word for word, idx in cv.vocabulary_.items() }
```

```python
idx_to_word
```

```python
from sklearn.model_selection import train_test_split

x_train, y_train = sparse_train_counts, train_df['label']
x_test, y_test = sparse_test_counts, test_df['label']
```

```python
#모델 훈련 (학습)
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=42, max_iter=10000, C=1)
logreg.fit(x_train, y_train)
logreg.score(x_train, y_train), logreg.score(x_test, y_test)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer

tv = TfidfVectorizer() #raw string -> tfidf
sparse_train_counts2 = tv.fit_transform(train_df['review'])
sparse_test_counts2 = tv.transform(test_df['review'])

x_train2, y_train2 = sparse_train_counts2, train_df['label']
x_test2, y_test2 = sparse_test_counts2, test_df['label']

from sklearn.linear_model import LogisticRegression

logreg2 = LogisticRegression(random_state=42, max_iter=10000, C=1)
logreg2.fit(x_train2, y_train2)
logreg2.score(x_train2, y_train2), logreg2.score(x_test2, y_test2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression

# 1. CountVectorizer로 단어 빈도 행렬 생성
cv = CountVectorizer()
train_counts = cv.fit_transform(train_df['review'])
test_counts = cv.transform(test_df['review'])

# 2. TfidfTransformer로 TF-IDF 변환
tt = TfidfTransformer()
sparse_train_counts2 = tt.fit_transform(train_counts)
sparse_test_counts2 = tt.transform(test_counts)

# 3. 로지스틱 회귀 모델 학습 및 평가
x_train2, y_train2 = sparse_train_counts2, train_df['label']
x_test2, y_test2 = sparse_test_counts2, test_df['label']

logreg2 = LogisticRegression(random_state=42, max_iter=10000, C=1)
logreg2.fit(x_train2, y_train2)

print("Train Accuracy:", logreg2.score(x_train2, y_train2))
print("Test Accuracy:", logreg2.score(x_test2, y_test2))
```
