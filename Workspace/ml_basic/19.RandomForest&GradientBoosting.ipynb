```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
x = df['data']
y = df['target']

y_label = df['target_names']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state= 777, stratify=y)
len(x_train), x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=500, max_depth=5, random_state=777)
model.fit(x_train, y_train)
model.score(x_train, y_train), model.score(x_test, y_test)
```

```python
for n, i in zip(df['feature_names'], model.feature_importances_):
    print(n, i)
```

```python
plt.barh(df['feature_names'], model.feature_importances_)
plt.show()
```

```python
#모델훈련
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model2 = GradientBoostingClassifier(n_estimators=1000, learning_rate=0.01, random_state=77777)
model2.fit(x_train, y_train)
model2.score(x_train, y_train), model2.score(x_test, y_test)
```

```python
for n, i in zip(df['feature_names'], model2.feature_importances_):
    print(n, i)
```

```python
plt.barh(df['feature_names'], model2.feature_importances_)
plt.show()
```
