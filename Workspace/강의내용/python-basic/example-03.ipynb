```python
import requests

headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9,ko;q=0.8,ko-KR;q=0.7"
}

url = "https://www.imdb.com/chart/top/"
response = requests.get(url, headers=headers)
```

```python
# response.content # binary text : b'....'
# response.text # text : '....' == response.content.decode('utf-8')
response.text[:300]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
titles = soup.select('.ipc-metadata-list .ipc-metadata-list-summary-item h3')
```

```python
len(titles)
```

```python
for title in titles:
    print(title.text)
```

```python
# <script type="application/ld+json"> 요소를 찾는 선택자
data_script = soup.select('script[type="application/ld+json"]')
```

```python
data_script
```

```python
data_script[0].text
```

```python
import json

json_data = json.loads(data_script[0].text)
```

```python
json_data
```

```python
print( len( json_data['itemListElement'] ) )
json_data['itemListElement'][0]
```

#### Web Crawling using Selenium

```python
# !pip install selenium
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
chrome = webdriver.Chrome()
```

```python
chrome.get("http://python.org") 
```

```python
menus = chrome.find_elements(By.CSS_SELECTOR, '#top ul.menu li')
pypi = None 
for m in menus: 
    if m.text == "PyPI": 
        pypi = m 
    print(m.text) 
```

```python
pypi.click()
```

```python
import time 
time.sleep(5) # 5초 대기 
chrome.quit() # 브라우저 종료 
```
