```python
# str은 series에 포함된 문자열 각각에 적용할 함수를 포함하는 객체
dong_filter = data['행정구역'].str.contains("오산시")
```

```python
data.iloc[:, 1:] = data.iloc[:, 1:].replace(',', '', regex=True)
```

```python
data.head()
```

```python
data.iloc[:, 1:] = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
```

```python
data.dtypes
```

```python
for col in data.columns[1:]:  # '행정구역' 제외
    data[col] = pd.to_numeric(data[col], errors='coerce')  # 숫자형으로 변환
```
