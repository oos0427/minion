<h1>Seaborn</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#     
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips)
sns.scatterplot(x= "total_bill", y = "tip", hue = "smoker", style= "time", data = tips)
plt.show()
```

```python
print(tips.describe(), '\n')
print(tips.info(), '\n')
print(type(tips))
```

```python
print(tips.groupby(["smoker", "sex", "time"])['tip'].mean(), '\n')
print(tips.groupby("time")['tip'].mean(), '\n')
print(tips.groupby("sex")['tip'].mean(), '\n')
```

```python
print(tips.groupby(["smoker", "sex", "time"])['tip'].size(), '\n')
print(tips.groupby("time")['tip'].size(), '\n')
print(tips.groupby("sex")['tip'].size(), '\n')
```

```python
print(tips.groupby(["smoker", "sex", "time"])['tip'].count(), '\n')
print(tips.groupby("time")['tip'].count(), '\n')
print(tips.groupby("sex")['tip'].count(), '\n')
```

```python
#    
#  
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
fmri.info()
```

```python
fmri.describe()
```

```python
#    
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
kind="line", data=fmri);
```

```python
fmri.loc[:, 'event'].unique()
```

```python
#            : wrap
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
col="subject", col_wrap=5, height=3, linewidth=2.5,
kind="line", data=fmri.query("region == 'frontal'"));
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#   : jitter=True
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, ax=ax1)  # jitter    
ax1.set_title('With Jitter')

#   : jitter=False
sns.stripplot(x="day", y="total_bill", data=tips, jitter=False, ax=ax2)  # jitter    
ax2.set_title('Without Jitter')

plt.show()
```

```python
#Bar plot
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic, palette="pastel")
```

```python
exercise = sns.load_dataset("exercise")
sns.catplot(x="time", y="pulse", hue="kind", col="diet", data=exercise)
```

```python
#   
x = np.random.normal(size=100)
sns.histplot(x, bins=20, kde=True) # kde : kernel density
plt.show()
#         
```

```python
#   
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
sns.jointplot(x="x", y="y", kind="kde", data=df) # kind : (scatter,hex,kde)
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species", markers=["o", "s", "^"]) # markers hue
#     hue      
plt.show()
```

```python
#   
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("R e g p l o t", pad = 20)
plt.show()
```

```python
#    
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.3)
g.add_legend()
```

```python
g = sns.FacetGrid(tips, col="day", height=4)
g.map(sns.barplot, "sex", "total_bill")
```

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
plt.show()
```
