```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from scipy import stats
```

```python
N = 10 # 데이터 갯수
K = 4 # 범주
mu_0 = np.ones(K) / K # 모든 요소의 값이 1인 nbarray를 만듬 -> 발생 확률
# print(mu_0)

np.random.seed(42) #항상 같은 난수를 형성
x = np.random.choice(K, N, p=mu_0)
```

```python
c = np.bincount(x)
```
